```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
sns.lineplot(x=df.index, y=df['SepalLengthCm'])
plt.title('Sepal Length by Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
sns.scatterplot(x=df['PetalLengthCm'], y=df['PetalWidthCm'],palette='rocket' , hue=df['Species'])
plt.title('Pental Length by Pental Width')
plt.xlabel('Pental Length (cm)')
plt.ylabel('Pental Width (cm)')
plt.show()
```

```python
sns.histplot(df['PetalLengthCm'], bins=50, kde=True, color="navy", edgecolor="black")
plt.title('Pental Length')
plt.xlabel('Pental Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x = df['Species'], y = df['PetalLengthCm'], data = df, palette="rocket")
plt.title('Species by Pental Length')
plt.xlabel('Species')
plt.ylabel('Pental Length (cm)')
plt.show()
```

```python
sns.barplot(x=df['SepalLengthCm'].mean() , y=df['Species'] , palette='rocket')
plt.title('Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Species')
plt.show()
```

```python
sns.pairplot(data=df, hue='Species', palette='rocket')
plt.show()
```
