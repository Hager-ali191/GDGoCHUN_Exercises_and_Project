**Student Registration System**


```python
# ask user for their name and age

print("Hello Student! , We would like to know more about you.")
user_name = input("What is your name?")
user_age = input("And your age?")
print("Good , now let’s get more info about you")


# ask user to enter three favorite subjects and store tjhem in a list
print("Want you to enter your favorite subjects , at most three.")
user_subject1= input("First Subject : ")
user_subject2= input("Second Subject : ")
user_subject3= input("Third Subject : ")

user_subjects = [user_subject1 , user_subject2 , user_subject3]

print(user_subjects)

# covert list of subjects info a tuple
user_subjects_tuple = tuple(user_subjects)
print(user_subjects_tuple)

# create dictionary to store the student’s :
        # name , subjects (tuple)
user_info = {
    'Name' : user_name,
    'Age' : user_age,
    'Favorite Subjecta' : user_subjects_tuple
}

print(user_info)

# print a welcome message using string concatenation including student name 
print(f"Welcome {user_name}!")
print("Welcome " + user_name + "!")

# repeat a welcome message two times using string concatenation
message = "Welcome " + user_name + "!" + '\n'
print(message * 2 )

# check if student is 18 or older and print "adult student" or "minor student"
if int(user_age) >= 18:
    print("Adult Student.")
else:
    print("Minor Student.")

# use loop to print each subject in the tuple 
for subject in user_subjects_tuple:
    print(subject)

# write a function greet() that print "Welcome to python learning!" and call it
def greet():
    print("Welcome to python learning!")
    
greet()
```
